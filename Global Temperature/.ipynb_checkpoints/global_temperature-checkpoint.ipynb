## How have Earth Surface temperatures changed over time?

```python
# Read in data into the dataframe
data = pd.read_csv('GlobalTemperatures.csv')
```

```python
data.head()
```

```python
data.tail()
```

```python
# list of attributes
list(data.columns.values)
```

```python
data.shape
```

```python
# consider LandAverageTemperature column
data_avg_temp = data.iloc[:,:2]
```

```python
data_avg_temp.describe()
```

```python
data_avg_temp.info()
```

```python
plt.figure(figsize = (15,5))
plt.title("Avg Land Temp 1750-2015")
plt.xlabel("Year")
plt.ylabel("Avg Land Temp")
plt.plot(data_avg_temp['LandAverageTemperature'])
plt.show()
```

```python
plt.figure(figsize = (15,5))
plt.title("Avg Land Temp 1750-2015")
plt.xlabel("Year")
plt.ylabel("Avg Land Temp")
plt.scatter(x=data_avg_temp['LandAverageTemperature'].index, y=data_avg_temp['LandAverageTemperature'])
plt.show()
```

```python
# convert to date time index
times = pd.DatetimeIndex(data_avg_temp['dt'])
```

```python
# group by year
grouped = data_avg_temp.groupby([times.year]).mean()
grouped.head()
```

```python
# Plot
plt.figure(figsize=(10,3))
plt.title("Yearly Average Land Temperature 1750-2015")
plt.xlabel("Years")
plt.ylabel("Yearly Average Land Temperature")
plt.plot(grouped['LandAverageTemperature'])
plt.show()
```

```python
data_avg_temp[times.year == 1750]
```

```python
data_avg_temp[np.isnan(data_avg_temp['LandAverageTemperature'])]
```

```python
# Use previous valid observation to fill gap

data_avg_temp['LandAverageTemperature'] = data_avg_temp['LandAverageTemperature'].fillna(method='ffill')
```

```python
# Regroup and plot
plt.figure(figsize=(10,3))
plt.title("Yearly Average Land Temperature 1750-2015")
plt.xlabel("Years")
plt.ylabel("Yearly Average Land Temperature")
plt.plot(grouped['LandAverageTemperature'])
plt.show()
```

### Model a Dataset

```python
from sklearn.linear_model import LinearRegression as LinReg
```

```python
x = grouped.index.values.reshape(-1,1)
y = grouped['LandAverageTemperature'].values
```

```python
reg = LinReg()
reg.fit(x,y)
y_preds = reg.predict(x)
print("Accuracy: " + str(reg.score(x,y)))
```

```python
plt.figure(figsize = (15,5))
plt.title("Linear Regression")
plt.scatter(x=x, y=y_preds)
plt.scatter(x=x, y=y, c="r")
```
