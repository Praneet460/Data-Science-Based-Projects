<h3>Subplotting</h3>

Subplotting is a technique for creating multiple plots that live side-by-side in one overall figure.

<h3>Working wiith Wine Review Dataset</h3>

```python
import pandas as pd

wine_review = pd.read_csv("winemag-data_first150k.csv", index_col = 0)
wine_review.head(5)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
# subplots takes two arguments
# first one controls number of rows
# second one controls number of columns
fig, axarr = plt.subplots(2,1, figsize=(12,8))
```

```python
axarr
```

To tell pandas which subplot we want a new plot to go in—the first one or the second one—we need to grab the proper axis out of the list and pass it into pandas via the ax parameter.

```python
fig, axarr = plt.subplots(2,1,figsize=(12,8))

wine_review['points'].value_counts().sort_index().plot.bar(
    ax=axarr[0], fontsize=12
)

wine_review['province'].value_counts().head(20).plot.bar(
    ax=axarr[1],fontsize=12
)
import seaborn as sns
sns.despine()
```

```python
fig, axarr = plt.subplots(2,2, figsize=(12,8))
```

```python
axarr
```

```python
fig, axarr = plt.subplots(2, 2, figsize=(12, 12))

wine_review['points'].value_counts().sort_index().plot.bar(
    ax=axarr[0][0], fontsize=12, color='mediumvioletred'
)
axarr[0][0].set_title("Wine Scores", fontsize=18)

wine_review['variety'].value_counts().head(20).plot.bar(
    ax=axarr[1][0], fontsize=12, color='mediumvioletred' )
axarr[1][0].set_title("Wine Varities", fontsize=18)

wine_review['province'].value_counts().head(20).plot.bar(
    ax=axarr[1][1], fontsize=12, color='mediumvioletred')
axarr[1][1].set_title("Wine Origins", fontsize=18)

wine_review['price'].value_counts().plot.hist(
    ax=axarr[0][1], fontsize=12, color='mediumvioletred')
axarr[0][1].set_title("Wine Prices", fontsize=18)

plt.subplots_adjust(hspace=.3)


sns.despine()

```

<h3>Working With Pokemon Datasset</h3>

```python
pokemon = pd.read_csv('pokemon.csv')
pokemon.head(4)
```

```python
fig, axarr = plt.subplots(2,1, figsize=(8,8))
sns.despine()
pokemon['attack'].plot.hist(
     ax=axarr[0], fontsize=12, color='mediumvioletred')
axarr[0].set_title("Pokemon Attack Rating", fontsize=18)

pokemon['defense'].plot.hist(
    ax=axarr[1], fontsize=12, color='mediumvioletred')
axarr[1].set_title("Pokemon Defense Rating", fontsize=18)
```
