<h1>Web Scrapping on DataCamp Instructors and Courses Site</h1>

Scrap the data and store it in a csv file. List of Instructor Names, Number of Students Enrolled, and Url of their profile is stored in a csv file name 'instructors_data_01.csv'. And the list of courses, Instructors Taught the courses ,and the link to the course is stored in another csv file named 'courses_data_01.csv'.Also created a function which list down the courses details taught by specific instructor just by giving the instructor name.

```python
# store url
url = 'https://www.datacamp.com/instructors?all=true'
```

```python
# import 'requests'
import requests

# make the request
r = requests.get(url)
# Extract HTML from Response object and print
#print(r.text)
```

```python
# Import BeautifulSoup from bs4
from bs4 import BeautifulSoup 
soup = BeautifulSoup(r.text, 'lxml')
# .instructor-block_description .instructor-block_link contains the link of the instructors profile

soup.select('.instructor-block__description .instructor-block__link ')
```

```python
# Generates the list of all the DataCamp instructors urls
instructor_urls = [url['href'] for url in soup.select('.instructor-block__description .instructor-block__link')]
instructor_urls = ['https://www.datacamp.com' + url for url in instructor_urls]
instructor_urls
```

```python
#Fetching all the instructor names
soup.select('.mb-sm')
```

```python
# Generate the list of all the datacamp instructor names
instructor_names = [name.text.strip() for name in soup.select('.mb-sm')]
instructor_names
```

```python
# Number of students enrolled in different instructors courses
soup.select('.pr-xs')
```

```python
#enrolled_students = [number.text for number in soup.select('.pr-xs') ]
import re
enrolled_students = [re.sub(',', '', re.findall('[\d,]+', str(i))[0]) for i in soup.select('.pr-xs')]
enrolled_students
```

```python
# Creating the dataframe
import pandas as pd
%matplotlib inline
instructor_df = pd.DataFrame({
    'Instructors name': instructor_names,
    'Profile': instructor_urls,
    'Students Enrolled':enrolled_students
})
print(instructor_df.shape)
instructor_df.head()
```

```python
# Store data into a csv file
instructor_df.to_csv('instructors_data.csv', index=False)
```

```python
# Store new url
new_url = 'https://www.datacamp.com/courses/all'
```

```python
# make the request
new_r = requests.get(new_url)
new_soup = BeautifulSoup(new_r.text, 'lxml')
# Scrap the List of all the courses Courses
new_soup.select('.course-block__title')
```

```python
# Generate the list of all the datacamp courses
course_names = [name.text.strip() for name in new_soup.select('.course-block__title')]
course_names
```

```python
# Scrap the list of all the course Instructors
course_instructors =  [name.text.strip() for name in new_soup.select('.course-block__author-name')]
course_instructors
```

```python
# Scrap the list of all the courses link
course_link =  [url['href'] for url in new_soup.select('.course-block__link')]
course_link = ['https://www.datacamp.com' + url for url in course_link]
course_link
```

```python
# Creating the dataframe
courses_df = pd.DataFrame({
    'Course Name': course_names,
    'Instructor': course_instructors,
    'Link':course_link
})
print(courses_df.shape)
courses_df.head()
```

```python
# Number of Courses by top 6 instructors
courses_df.Instructor.value_counts().head(6)
```

```python
# have a look on bar plot
import matplotlib.pyplot as plt
import seaborn as sns
ax = courses_df.Instructor.value_counts().head(10).plot.bar(figsize=(10,8), fontsize=18, color='mediumvioletred')
ax.set_title("Number of courses Offered", fontsize = 25)
sns.despine()
```

```python
# Store it into a csv file
courses_df.to_csv('courses_data_01.csv', index=False)
```

<h3>Function created to show the courses taught by specific instructor</h3>

```python
def instructor_name(input):
    
    courses_list = courses_df[courses_df.Instructor == input]
    return courses_list
```
